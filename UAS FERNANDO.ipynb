```python
df = pd.read_csv("students.csv")
df
```

```python
gpa_columns = ['Sem1 GPA', 'Sem2 GPA', 'Sem3 GPA', 'Sem4 GPA', 
               'Sem5 GPA', 'Sem6 GPA', 'Sem7 GPA', 'Sem8 GPA']
df['Total_Skor'] = df[gpa_columns].sum(axis=1)
```

```python
plt.figure(figsize=(14,6))
plt.bar(np.arange(1, 51), df['Total_Skor'][:50], color='skyblue')
plt.title("Total Skor Akademik (50 Mahasiswa Pertama)")
plt.xlabel("Index Mahasiswa")
plt.ylabel("Total Skor")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,8))
for col in ['Sem1 GPA', 'Sem4 GPA', 'Sem8 GPA']:
    plt.plot(df.index[:50], df[col][:50], marker='o', label=col)
plt.title("Distribusi GPA Semester (50 Mahasiswa Pertama)")
plt.xlabel("Index Mahasiswa")
plt.ylabel("Nilai GPA")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Contoh data (ganti dengan data kamu sendiri)
X = np.array([[1], [2], [3], [4], [5]])  # Input (fitur), harus 2D
y = np.array([60, 65, 70, 75, 80])       # Target, 1D

# Membuat dan melatih model
model = LinearRegression()
model.fit(X, y)

# Melakukan prediksi untuk mahasiswa ke-1000
prediksi = model.predict([[1000]])[0]
print(f"Prediksi Total Skor Mahasiswa ke-1000: {prediksi:,.2f}")
```
